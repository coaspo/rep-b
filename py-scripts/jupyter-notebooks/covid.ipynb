```python
import math
import numpy
import scipy
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def func(x, a, b, c):
    return a * numpy.exp(b * x) + c


def exponential_fit(y):
    x=numpy.asarray([*range(0, len(y))])
    a=scipy.optimize.curve_fit(func, x, y)
    pfit, perr = curve_fit(func, x, y)
    print(pfit)
    doubling_days = "{:.1f}".format(math.log(2)/pfit[1])
    print ('doubles every ', doubling_days, 'days.')

    plt.plot(x, y, "ro", label = "data")
    y2 = func(x, *pfit)
    plt.plot(x, y2, "b", label = "fit")
    plt.legend()
    plt.show()
    
#1/31/2020 - lastest count.
china_counts = numpy.array([571,830,1287,1975,2744,4515,5974,7711,9692,11791])
exponential_fit(china_counts)
```
